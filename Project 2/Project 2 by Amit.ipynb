```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt 
log=linear_model.LogisticRegression()
train=pd.read_csv('train.csv')
test_x=pd.read_csv('test_X.csv')
train=train.replace(['male', 'female'], ['0','1'])
train=train.drop(columns=['Ticket','Cabin','Name','PassengerId','Fare','Embarked'])
train=train.fillna(0)
test_x=test_x.replace(['male', 'female'], ['0','1'])
test_x=test_x.drop(columns=['Ticket','Cabin','Name','PassengerId','Fare','Embarked'])
test_x=test_x.fillna(0)
test_y=pd.read_csv('test_Y.csv')
test_y=test_y.drop(columns=['PassengerId'])
train_x=train.drop(['Survived'],axis=1)
train_y=train['Survived']
log.fit(train_x,train_y)
pred = log.predict(test_x)
accuracy_score(test_y,pred)
```

```python
plt.plot(test_y,color='black', linewidth=2)
plt.plot(pred, color='red', linewidth=2)

plt.xticks(())
plt.yticks(())

plt.show()
```
